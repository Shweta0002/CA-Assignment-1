```python
import numpy as np
import pandas as pd
import sklearn 
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

```python
dataset = pd.read_csv("521(1).csv")
dataset= dataset.iloc[1000:]
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
y=dataset['CPI'] 

corr_matrix = dataset.corr().abs()


upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
print(to_drop)
```

```python
instructions = dataset['instructions:u'].values.reshape(-1,1) 

y=dataset['CPI']  
X = dataset.drop(['CPI','cycles:u','instructions:u','time'], axis = 1) 
#X = dataset.drop(['CPI','cycles:u','time','itlb_misses.walk_pending:u','iTLB-load-misses','L1-icache-load-misses','cache-misses:u','node-load-misses','node-store-misses','l2_rqsts.swpf_miss:u','branch-load-misses','dtlb_load_misses.walk_pending:u','icache_64b.iftag_miss:u','l2_rqsts.code_rd_miss:u','instructions:u','page-faults:u','offcore_requests.l3_miss_demand_data_rd:u','l2_rqsts.all_demand_miss:u', 'dTLB-store-misses','LLC-load-misses','L1-dcache-load-misses','dtlb_load_misses.walk_pending:u.1', 'IPC'], axis = 1) 
X = np.divide(X, instructions) 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.1, random_state=0) 
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor_linear = LinearRegression()
# temp=regressor.fit(X_train, y_train)
temp=regressor_linear.fit(X_train, y_train)
y_pred_linear = regressor_linear.predict(X_test)
```

```python
print('Linear Intercept term: ',temp.intercept_)
print('Linear Coefficients: \n',temp.coef_) 
coeff_df = pd.DataFrame(temp.coef_, X_train.columns, columns=['Cofficient'])  
print(coeff_df)
print()

# print(temp.coef_)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from math import sqrt
print()


rsq = r2_score(y_test,y_pred_linear) 
print('R-square: ',rsq)
rmse = sqrt(mean_squared_error(y_test, y_pred_linear))  # Computing value of Root Mean Square Error for testing data
print("RMSE:- ",rmse)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred_linear)
plt.xlabel("Actual : $Y_i$")
plt.ylabel("Predicted : $\hat{Y}_i$")
plt.title("Predicted  vs Actual: For Linear Regression")
plt.show()
# print(y_pred_linear)
# print(y_pred)
```
